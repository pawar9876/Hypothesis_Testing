CHI-SQUARE TEST
Association between Device Type and Customer Satisfaction
Background:
Mizzare Corporation has collected data on customer satisfaction levels for two types of smart home devices: Smart Thermostats and Smart Lights. They want to determine if there's a significant association between the type of device purchased and the customer's satisfaction level.
Data Provided:
The data is summarized in a contingency table showing the counts of customers in each satisfaction level for both types of devices:
Satisfaction	Smart Thermostat	Smart Light	Total
Very Satisfied	50	70	120
Satisfied	80	100	180
Neutral	60	90	150
Unsatisfied	30	50	80
Very Unsatisfied	20	50	70
Total	240	360	600
Objective:
To use the Chi-Square test for independence to determine if there's a significant association between the type of smart home device purchased (Smart Thermostats vs. Smart Lights) and the customer satisfaction level.
Assignment Tasks:
1. State the Hypotheses:
2. Compute the Chi-Square Statistic:
3. Determine the Critical Value:
Using the significance level (alpha) of 0.05 and the degrees of freedom (which is the number of categories minus 1)
4. Make a Decision:
Compare the Chi-Square statistic with the critical value to decide whether to reject the null hypothesis.

```python
# importing libraries for data analyzing and visualization
import pandas as pd
import numpy as np
from scipy import stats
from scipy.stats import chi2_contingency
```

```python
data=pd.DataFrame({'Smart Thermostat':[50,80,60,30,20],'Smart Light':[70,100,90,50,50]})
data  # create data
```

```python
# set in index satisfied type
data.index=['Very Satisfied','Satisfied','Neutral','Unsatisfied','Very Unsatisfied']
data
```

# Hypothesis State
HO null hypothesis="there is not sognificant association between customer
satisfaction level"


H1 alternative hypothesis=""there is  sognificant association between customer satisfaction level"

```python
#HO null hypothesis="there is not sognificant association between customer satisfaction level"
#H1 alternative hypothesis=""there is  sognificant association between customer satisfaction level""
```

```python
data.plot(kind='bar',    # comparing Smart Thermostat vs Smart Light satisfied level
        stacked=False)
```

```python
chi2,p,dof,exp=chi2_contingency(data) # test statistic, p-value,df,expected frequency
```

# Computing chi2 statistic value

```python
chi2
```

# Determine critical value and degrees of freedom

```python
print("critical_value",np.round(p,2))  # here critical value 0.23
print("dof",dof)                        # degrees of freedom is 4
```

```python
exp  # expected frequency
```

# Decision Making

```python
if p<0.05:
    print("Reject Null Hypothesis ")
    print("there's  significant association between the type of device purchased and the customer's satisfaction level.")

else:
    print("Accept Null Hypothesis H0")
    print("there's a  not significant association between the type of device purchased and the customer's satisfaction level.")
```

To use the Chi-Square test for independence to determine if there's a significant association between the type of smart home device purchased (Smart Thermostats vs. Smart Lights) and the customer satisfaction level when p (critical_value) is smaller  than alpha value .

but we got p> alpha value so we accept the null hypothesis it says there is not significant association between the type of smart home devices and customer satisfaction level.
